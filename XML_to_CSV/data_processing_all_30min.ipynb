```python
%load_ext autotime

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

from tqdm import trange,tqdm
warnings.filterwarnings('ignore')
```

```python
def data_processing(data, common_cols):
    data = data.rename(columns={'Unnamed: 0': 'time'})
    
    data['basel_value'].fillna(value = 0, inplace = True)
    data['tb_value'].fillna(value = 0, inplace = True)

    data['basel_value'] = data['basel_value']+data['tb_value']
    data.drop(columns = ['tb_value'], inplace = True)
    
    for i in range(len(data)):
        if (data['basel_value'][i] == 0):
            data['basel_value'][i] = np.nan
            
    data['basel_value'] = data['basel_value'].ffill()
    data['basel_value'] = data['basel_value'].bfill()
    
    possible_cont_cols = ['gl_value', 'fs_value', 'basis_heart_rate', 'basis_gsr', 'basis_skin_temp', 'basis_air_temp']
    fillable_cols = [col for col in data.columns if col in possible_cont_cols]
    data[fillable_cols] = data[fillable_cols].ffill()
    data[fillable_cols] = data[fillable_cols].bfill()
#     interpolate_cols = [col for col in data.columns if col in possible_cont_cols]
#     data[interpolate_cols] = data[interpolate_cols].interpolate(method = 'linear', limit_direction = 'both')
    data.fillna(value = 0, inplace = True)
    
    data['gl_predict'] = np.nan
    data['gl_predict'][0:len(data)-1] = data['gl_value'][1:len(data)]
#     data.drop(columns = ['time'], inplace = True)
#     data = data[common_cols]
    data = data[:-1]
    
    return (data)
```

```python
odd_files = ['570-ws-training.csv', '575-ws-training.csv', '591-ws-training.csv', '588-ws-training.csv', '559-ws-training.csv', '563-ws-training.csv']
common_cols_9 = ['meal_type_Lunch', 'basis_skin_temp', 'basis_gsr', 'meal_carbs', 'meal_type_Breakfast', 'bolus_dose', 'meal_type_Dinner', 'basel_value', 'fs_value', 'gl_value', 'gl_predict']
common_cols_19 = ['meal_type_Lunch', 'basis_skin_temp', 'basis_gsr', 'meal_carbs', 'bwz_carb_input', 'meal_type_Breakfast', 'basis_steps', 'exercise_duration', 'basis_air_temp', 'meal_type_HypoCorrection', 'basis_heart_rate', 'exercise_intensity', 'basis_sleep_quality', 'bolus_dose', 'meal_type_Dinner', 'sleep_quality', 'meal_type_Snack', 'basel_value', 'fs_value', 'gl_value', 'gl_predict']

for file in os.listdir('../data/derived/30_min_no_interpolate/'):
    print (file)
    data = pd.read_csv('../data/derived/30_min_no_interpolate/'+file)
    if file in odd_files:
        final_data = data_processing(data, common_cols_19)
    else:
        final_data = data_processing(data, common_cols_9)
        
    final_data.to_csv('../data/derived/30_min_complete_dataset/'+file, index = False)
```
